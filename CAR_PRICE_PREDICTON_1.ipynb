```python
df=cardata.copy();
```

```python
# removing non year values form df
df=df[df['year'].str.isnumeric()]
```

```python
# changing datatype of year from object to int
df['year'] = df['year'].astype(int)
```

```python
df.shape
```

```python
l=len(df[df['Price']=='Ask For Price'])
```

```python
l/len(df)*100 # percentage of non price values in price column
# as percentage is less so we will delete these rows from the dataset
```

```python
df=df[df['Price']!='Ask For Price']; # removing non price rows
```

```python
df.shape
```

```python
def deletecomma(str):  # function to delete commmma in price
  str=str.replace(',','');
  return str
```

```python
print(deletecomma('0,000,0,0'))
```

```python
df['Price'] = df['Price'].apply(deletecomma) # applying function on price column
```

```python
df['Price']=df['Price'].astype(int) # changing datatype of price from object to int
```

```python
df.info()
```

```python
df.head()
```

```python
# function to remove kms from the kms_drive column
def deletekms(str):
  str=str.split(' ')[0]
  return str
```

```python
df['kms_driven']=df['kms_driven'].apply(deletekms)
```

```python
df['kms_driven']=df['kms_driven'].apply(deletecomma) # deleting commmma from kms_Driven column
```

```python
#keeping only first three words from name of the company for further proceedings
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
# converting kms driven column to int type
df['kms_driven']=df['kms_driven'].astype(int) # it is done later
```

```python
df.head()
```

```python
df.info()
```

```python
df[df['fuel_type'].isnull()]
```

```python
#replacing their fuel type with most frequent fuel type
from sklearn.impute import SimpleImputer
imp=SimpleImputer(strategy='most_frequent')
imp.fit_transform(df[['fuel_type']]);
```

```python
imp.statistics_[0] # frequent fuel type
```

```python
df['fuel_type'].fillna(imp.statistics_[0],inplace=True)  # replacing null values of fuel_type with most frequent fuel type
```
